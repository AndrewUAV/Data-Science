Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python

grouped_by_position = python_data.groupby('Посада')
group_names = grouped_by_position.groups.keys()
group_names
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада",
виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
python_data_new = python_data.groupby(['Посада'])['Зарплата на місяць'].agg(['min', 'max'])
python_data_new
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return row.mean()

python_data_new['avg'] = python_data_new.apply(fill_avg_salary, axis=1)
python_data_new
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
description = python_data_new['avg'].describe()
description
```

Збережіть отриману таблицю в CSV файл

```python
python_data_new.to_csv('python_data_new.csv', index=False)
```

```python
plt.figure(figsize=(8, 4))
plt.hist(python_data_new['avg'], bins=10, color='red', edgecolor='black')
plt.xlabel('Середнє значення')
plt.ylabel('Частота')
plt.title('Гістограма середніх значень')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=python_data_new['avg'], color='lightblue')
plt.ylabel('Середнє значення')
plt.title('Ящик для середніх значень')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(python_data_new['avg'], fill=True, color='purple')
plt.xlabel('Середнє значення')
plt.ylabel('Щільність')
plt.title('Графік щільності середніх значень')
plt.grid(True)
plt.show()
```
